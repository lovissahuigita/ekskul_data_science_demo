```python
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.dates as mdates
import datetime as dt
```

```python
aapl = pd.read_csv("aapl.csv")
aapl = aapl.set_index("Date").sort_index()  # add sort_index
```

```python
# Data Visualisasi
plt.plot(aapl.index, aapl.Close, label="AAPL")

plt.title("AAPL Prices")
plt.xlabel("Date")
plt.ylabel("Close Price")

# fix intervals
fmt = mdates.MonthLocator(interval=1)
plt.gca().xaxis.set_major_locator(fmt)
plt.xlim([aapl.first_valid_index(), aapl.index[-1]])

plt.show()
```

```python
jpm = pd.read_csv("jpm.csv")
jpm = jpm.set_index("Date").sort_index()  # add sort_index
```

```python
# Data Visualisasi 2 stock
plt.plot(aapl.index, aapl.Close, label="AAPL")
plt.plot(jpm.index, jpm.Close, label="JPM")

plt.title("AAPL & JPM Prices")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()

# fix intervals
fmt = mdates.MonthLocator(interval=1)
plt.gca().xaxis.set_major_locator(fmt)
plt.xlim([aapl.first_valid_index(), aapl.index[-1]])

plt.show()
```

```python
# Data Visualisasi portfolio
plt.plot(aapl.index, aapl.Close, label="AAPL")
plt.plot(jpm.index, jpm.Close, label="JPM")

portfolio = aapl.Close * 0.1 + jpm.Close * 0.9
plt.plot(jpm.index, portfolio, label="Portfolio")

plt.title("AAPL & JPM Prices")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()

# fix intervals
fmt = mdates.MonthLocator(interval=1)
plt.gca().xaxis.set_major_locator(fmt)
plt.xlim([aapl.first_valid_index(), aapl.index[-1]])

plt.show()
```

```python
# Data Visualisasi portfolio
# plt.plot(aapl.index, aapl.Close, label="AAPL")
# plt.plot(jpm.index, jpm.Close, label="JPM")

portfolio = aapl.Close * 0.1 + jpm.Close * 0.9
plt.plot(jpm.index, portfolio, label="Portfolio", color='lightgrey')

rolling_mean_portfolio = portfolio.rolling(window=14).mean()
plt.plot(jpm.index, rolling_mean_portfolio, label="Rolling Mean")

std = rolling_mean_portfolio.std()
plt.plot(jpm.index, rolling_mean_portfolio + std, label="Standard Deviation", color="green")
plt.plot(jpm.index, rolling_mean_portfolio - std, label="Standard Deviation", color="green")

plt.title("Lovissa's Portfolio Bollinger Band")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()

# fix intervals
fmt = mdates.MonthLocator(interval=1)
plt.gca().xaxis.set_major_locator(fmt)
plt.xlim([aapl.first_valid_index(), aapl.index[-1]])

plt.show()
```
